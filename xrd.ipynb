```python
from pymatgen.ext.matproj import MPRester
from pymatgen.analysis.diffraction.xrd import XRDCalculator

from matplotlib import pyplot as plt
```

```python
with MPRester("u0QTXogbkTqal3Dr") as m:
    structure = m.get_structure_by_material_id("mp-13314")
```

```python
xrd = XRDCalculator("CuKa1")
xrd_pattern = xrd.get_pattern(structure)
xrd.get_plot(structure, two_theta_range=(0, 25), annotate_peaks=True)
plt.show()
hkls = xrd_pattern.hkls
```

```python
print("Premier pic :", hkls[0])
print("Angle 2 entre les 2 rayons :", xrd_pattern.x[0])
print("Intensit du pic :", xrd_pattern.y[0])
```

```python
print("Deuxime pic : ", hkls[1])
print("Angle 2 entre les 2 rayons :", xrd_pattern.x[1])
print("Intensit du pic :", xrd_pattern.y[1])
```
